- Since we are only interested to know the rainfall of months june and july and we want to find out the probability of flood in a partiular year we are taking the corresponding features only

- probbability of flooding in a year given that it rained more than a threshold(500mm) in these months.We required to create more features in order to find these probabilities

```python
flood_kerala["JUN_GT_500"]=(flood_kerala["JUN"]>500).astype("int")
flood_kerala["JUL_GT_500"]=(flood_kerala["JUL"]>500).astype("int")
```

```python
df_small = flood_kerala.loc[:,['YEAR','JUN_GT_500','JUL_GT_500','FLOODS']]
```

## Encoding Data

```python
from sklearn import preprocessing
```

```python
label_enc=preprocessing.LabelEncoder()
```

```python
df['FLOODS']=label_enc.fit_transform(df['FLOODS'])
```

```python
x=df.drop(columns=["SUBDIVISION","FLOODS"])
y=df['FLOODS']
```

## Spliting The Data for Testing And Trainng

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42 )
```

```python
x_train
```

```python
y_train
```

## Creating Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
logicreg=LogisticRegression()
```

```python
logicreg
```

```python
logicreg.fit(x_train,y_train)
```

```python
y_predict=logicreg.predict(x_test)
```

```python
y_predict
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
acc=accuracy_score(y_test,y_predict)
```

```python
acc
```

```python
con=confusion_matrix(y_test,y_predict)
```

```python
con
```

```python
f1=f1_score(y_test,y_predict)
```

```python
f1
```

```python
pre=precision_score(y_test,y_predict)
```

```python
pre
```

```python
recall=recall_score(y_test,y_predict)
```
